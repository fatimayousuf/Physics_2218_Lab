# Preliminary Data Analysis
This notebook is intended to explore ways we can analyze the circuit data for our experiment. I will first be working with idealized data from the circuit simulator. I will then add noise to the simulation data and check if I can get  results for the min/maxes similar to the idealized data. The data I am using here is from a simulated RLC circuit.

```python
import pandas as pd 
import numpy as np
from scipy.signal import argrelextrema

# Get the CSV data
filepath = '/Users/fatimayousuf/Downloads/RLC_Circuit.csv'

# Read the csv file and organize data into a pandas dataframe
df = pd.read_csv(filepath)
df.columns = ['t', 'v']
df = df.iloc[1:] # not sure why, by the first row is just (t=0, v=0), so we can get rid of that

# Separate time and voltage data into lists
t = df[df.columns[0]].to_list()
v = df[df.columns[1]].to_list()
```

```python
import matplotlib.pyplot as plt

# Plot the collected data
plt.plot(t, v)
plt.ylabel('Voltage on Capacitor')
plt.xlabel('Time')
plt.show()
```

```python
# Find local maxes and mins. Append results to the dataframe by creating "min" and "max" columns
df['min'] = df.v[(df.v.shift(1) > df.v) & (df.v.shift(-1) > df.v)]
df['max'] = df.v[(df.v.shift(1) < df.v) & (df.v.shift(-1) < df.v)]

# Plot results. Maxes are colored green and mins red.
plt.scatter(df['t'], df['min'], c='r')
plt.scatter(df['t'], df['max'], c='g')
plt.plot(t, v)
plt.ylabel('Voltage on Capacitor')
plt.xlabel('Time')
plt.show()
```

```python
# Print the values of the points of attraction
local_mins = df[~df['min'].isnull()]
print("Local Mins: ")
print(local_mins[['t', 'v']])

local_maxes = df[~df['max'].isnull()]
print("\nLocal Maxes:")
print(local_maxes[['t', 'v']])
```

### Adding Noise 

The oscilloscope data will be very noisy, so I will add noise to the dataset and see if I can calculate similar values for the local mins and maxes.

```python
# Parameters for creating noise
np.random.seed(100)
mu, sigma = 0, 0.1 

# Create a Gaussian noise matrix with the same dimensions as the dataset
noise = np.random.normal(mu, sigma, [len(df.v),1]) 
zeroes = np.zeros((len(df.v),1))
noise = np.hstack([zeroes, noise, zeroes, zeroes]) # we only want to add noise to the voltage column

# Add noise to the data
df_noisy = df + noise

# Plot the noisy data
v_noisy = df_noisy[df_noisy.columns[1]].to_list()
plt.plot(t, v_noisy)
plt.show()
```

```python
# Calculate local mins and maxes of noisy data
n=int(1e4) # number of points to be checked before and after 

# Find local peaks (for quicker computation, I will only be looking at every tenth value as there are 100,000 rows)
df_noisy = df_noisy.iloc[::10]
df_noisy['min'] = df_noisy.iloc[argrelextrema(df_noisy.v.values, np.less_equal, order=n)[0]]['v']
df_noisy['max'] = df_noisy.iloc[argrelextrema(df_noisy.v.values, np.greater_equal, order=n)[0]]['v']

# Plot the collected data
plt.plot(t, v_noisy, zorder = 0)
plt.scatter(df_noisy['t'], df_noisy['min'], c='r', zorder = 100)
plt.scatter(df_noisy['t'], df_noisy['max'], c='g', zorder = 200)
plt.ylabel('Voltage on Capacitor')
plt.xlabel('Time')
plt.show()
```

```python
# Print the values of the points of attraction
local_mins = df_noisy[~df_noisy['min'].isnull()]
print("Local Mins: ")
print(local_mins[['t', 'v']])

local_maxes = df_noisy[~df_noisy['max'].isnull()]
print("\nLocal Maxes:")
print(local_maxes[['t', 'v']])
```

The calculated local mins and maxes on the noisy data are quite similar to the results from the idealized data! Even if we just look at the graph above, the calculated values seem to be fairly accurate.
