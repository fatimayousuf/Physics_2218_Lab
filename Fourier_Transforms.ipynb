### Fourier Transforms
I have been doing some background research and I came upon Fourier transformations. I thought it might be interesting to do a fast fourier transform on our data and see what comes up. First I'll try doing a transform on a $sin(2 \pi x)$ wave. The frequency of this wave is $2 \pi /2\pi =1$, so it should be pretty easy to tell if I am doing this right. Then I'll apply this to some simulation data in another notebook (I think it will be "Calculating Attractors".

```python
import numpy as np
import matplotlib.pyplot as plt

time = np.arange(0, 10.1, 0.1)
samplingFrequency = 1/0.1
amplitude = np.sin(2*np.pi*time)
plt.plot(time, amplitude)

# Create a Fourier spectrum (Reference: https://pythontic.com/visualization/signals/fouriertransform_fft)
fourierTransform = np.fft.fft(amplitude)/len(amplitude)           # Normalize amplitude
fourierTransform = fourierTransform[range(int(len(amplitude)/2))] # Exclude sampling frequency
tpCount     = len(amplitude)
values      = np.arange(int(tpCount/2))
timePeriod  = tpCount/samplingFrequency
frequencies = values/timePeriod

# Plot the spectrum
fig, ax = plt.subplots()
ax.plot(frequencies, np.abs(fourierTransform))
plt.show()
```

Looks like it works! There is a spike around frequency = 1
